```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress 
```

```python
cities=pd.read_csv('cities.csv')

cities.head()
```

```python
cities.to_html("newhtml.html", index=False, classes=["table","table-striped","table-hover"])
```

```python
plt.scatter(x=cities['Lat'], y=cities['Max Temp'])
plt.xlabel('Lat')
plt.ylabel('Max Temperature')
plt.grid()
plt.title('City Latitude vs. Max Temperature (F)')
plt.savefig('Images/fig1.png')
plt.show()
```

```python
plt.scatter(x=cities['Lat'], y=cities['Humidity'])
plt.xlabel('Lat')
plt.ylabel('Humidity')
plt.title('City Latitude vs. Humidity (%)')
plt.grid()
plt.savefig('Images/fig2.png')
plt.show()
```

```python
plt.scatter(x=cities['Lat'], y=cities['Cloudiness'])
plt.xlabel('Lat')
plt.ylabel('Cloudiness')
plt.title('City Latitude vs. Cloudiness (%)')
plt.grid()
plt.savefig('Images/fig3.png')
plt.show()
```

```python
plt.scatter(x=cities['Lat'], y=cities['Wind Speed'])
plt.xlabel('Lat')
plt.ylabel('Wind Speed')
plt.title('City Latitude vs. Wind Speed (mph)')
plt.grid()
plt.savefig('Images/fig4.png')
plt.show()
```
